**Data grouped Representation**

Must use groupby with aggregation function. otherwise it will give object.

````python
df.head()
````

````python
df.groupby('Country') # It will Give the object type
````

````python
df.groupby('Country').sum()
````

````python
df.groupby('Country',).min()
````

````python
df.groupby(['Country', 'Region']).min()
````

````python
# Groupby based on multiple column

df.groupby(['Country', 'Region'])[['Total', 'Quantity']].sum()
````

````python
# concerting str to Timestamp type
df['Date of Purchase'] = pd.to_datetime(df['Date of Purchase'])
````

````python
# Dynamically creating new column

df.groupby(df['Date of Purchase'].dt.year)['Quantity'].min()
````
