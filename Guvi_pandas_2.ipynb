**Few important statistical functions Using pandas**

1.  Sum
2.  Min
3.  Max
4.  Mean
5.  Variance
6.  Percentage
7.  Generating new column
8.  Correlation
9.  Absolute

````python
print(df['Total'].min())
print(df['Total'].max())
print(df['Total'].sum())
print(df['Total'].mean())
````

````python
# We can use Category column also (Through Ascending and Descending order)

print(df['Sales Person'].max())
print(df['Sales Person'].min())

print(df['Region'].min())
print(df['Country'].max())
````

````python
df[df['Total']==df["Total"].min()]
````

````python
# Finding ratio between selling and quantity

df['Selling Ratio'] = df['Total'] / df['Quantity']
df.head()
````

````python
df[df['Selling Ratio'] ==  df['Selling Ratio'].min()]
````

````python
df[df['Selling Ratio'] ==  df['Selling Ratio'].max()]
````

````python
# I want to give 10% Bonus from selling price

df['Bonus(10%)'] = df['Total'] / 10
df.head()
````

correlation

````python
df.corr()
````

````python
df[['Total','Selling Ratio']].corr()
````
